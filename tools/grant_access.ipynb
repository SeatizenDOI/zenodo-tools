The first step is to go to zenodo and add the grant for the person one time.






In the network tab, check the grants request to get the user id. You will need it.

```python
import json
import requests

USER_ID = "621279"
ACCESS_TOKEN = ""

params = {'access_token': ACCESS_TOKEN}
headers = {"Content-Type": "application/json"}

# Here a list of doi.
dois = [15192770, 15192767, 15192763, 15189300, 15189271, 15189250, 15189219, 15189192, 15189147, 15189133, 15189114, 15189094, 15189057, 15189021, 15188988, 15188908, 15188879, 15188866, 15188861, 15188858, 15188853, 15188842, 15188835, 15188830, 15188824, 15188807, 15188790, 15188776, 15188734, 15188731, 15188686, 15188662, 15188619, 15188584, 15188569, 15188538, 15188517, 15188492, 15188474, 15188469, 15188460, 15188453, 15188447, 15188444, 15188437, 15188432, 15188428, 15188394, 15188384, 15188366, 15188333, 15188300, 15188275, 15188252, 15188227, 15188175, 15188150, 15188135, 15188111, 15188108, 15188103, 15188098, 15188095, 15188085, 15188080, 15188075, 15188070, 15188065]


for i, doi in enumerate(dois):
    print(f"{i}/{len(dois)}", doi)
    data = {"grants":[{"subject":{"type":"user","id":USER_ID},"permission":"manage","notify":True}]}
    request = f"https://zenodo.org/api/records/{doi}"
    print(request)
    r = requests.get(f"{request}/access/users",params=params, headers=headers)
    print(r.json())
    r = requests.post(f"{request}/access/grants", params=params, data=json.dumps(data), headers=headers)

    print(r.status_code, r.json())
    print()
```
